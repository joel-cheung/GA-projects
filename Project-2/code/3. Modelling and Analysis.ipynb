# Background

Like everyone, buyers and sellers would want to negotiate and buy/sell their property for a price that is deemed reasonable by the market (as observed from past SalePrice), and hopefully minimize their losses (if the owners are desperate), and if possible, make a little profit off the transaction.

# Problem Statement

Given the self-interest of both the buyer and seller (or through their agents), there are many opportunity for people to capitalize on the asymmetric information of the other party by exploiting them in the form of under/over quoting them. To address this issue, my data-driven team of property agents developed a pricing model so that we can better advice our clients on the SalePrice they can expect to get in a fair trade, and the model's predictions can act as benchmarks to be used in negotiations. With the opening up borders and inter-state travelling returning, it is expected that there will be a greater need for our service and we will not only predict the expected house prices, but also the key areas of improvement homeowners can engage in to increase their chances of fetching a higher price when selling.

As a starting point, my team has identified the city of Ames in Iowa to be the first city to embark on this project.

Based on the challenge information, the metric of interest of the models will be Root Mean Square Error (RMSE).

# Importing Libraries

```python
# Usual libraries
import pandas as pd
import numpy as np

# For visualization and EDA
import seaborn as sns
import matplotlib.pyplot as plt

# Machine Learning
from sklearn.linear_model import Ridge, Lasso, LinearRegression, RidgeCV, LassoCV
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn import metrics

# For preprocessing
from sklearn.preprocessing import StandardScaler

import pickle

import warnings
warnings.filterwarnings('ignore')
```

# Importing Data

```python
# Importing both the training dataset
data = pd.read_csv("../data/data_v2.csv")
```

```python
# Importing original list of numerical variables (this includes SalePrice)
numerical_var = pickle.load(open("../data/numerical_variables.pkl", "rb"))
```

```python
# Removing SalePrice from list to retrieve only numerical explanatory variables
numerical_var.remove('SalePrice')
```

# Creating Functions to be used later

### Creating Formula to Retrieve Numeric and Categorical Variables

This is done as a succinct way to retrieve categorical and numerical variables in the dataset.

```python
def get_cat_var(data):
    cat_var = []
    for feature in data:
        if data[feature].dtype == object:
            cat_var.append(feature)
        else:
            pass
    return cat_var
```

```python
def get_num_var(data):
    num_var = []
    for feature in data:
        if data[feature].dtype != object:
            num_var.append(feature)
        else:
            pass
    return num_var
```

# Train/Test Split

This is done before standard scaling to prevent data leakage.

```python
y = data['SalePrice']
X = data.loc[:, data.columns != 'SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 42)
```

# StandardScaling Variables

This is done to ensure stability amongst the variables, where the magnitude of each variable will not disproportionately influence their coefficient sizes.

In the previous notebook, the categorical variables have been one-hot-encoded, with one of their levels removed to avoid perfect multicollinearity. 

Without further preprocessing, the standard scaler will be applied to all numerical variables, and this include the binarized categorical variable, and that will not be appropriate. Thus, there is a need to separate those binary columns from the numerical explanatory variables.

This is the reason for picking the original list of numeric explanatory variables from the first notebook, to be used to separate the binarized columns.

```python
# Instantiating standard scaler
scaler = StandardScaler()

# Fitting scaler on train set and applying scaler on both X_train and X_test
X_train[numerical_var] = scaler.fit_transform(X_train[numerical_var])
X_test[numerical_var] = scaler.transform(X_test[numerical_var])
```

The reason for fitting the scaler on the train set and not having seperate scalers is due to data leakage - the phenomenon where information from outside the training dataset is used to create the model.

Since the test set is used to mimic the unseen _**test**_ dataset (in this context, the kaggle test dataset), any information about the distribution of the _**test**_ dataset should not be captured in the model.

```python
X_train.head()
```

```python
X_test.head()
```

Note that the numerical variables are scaled, with the binarized columns remaining as 1s and 0s. It is okay for the numerical columns to take on negative values as the scale in question is one similar to a normal distribution, i.e. any values left of the mean is negative.

With the train and test datasets properly preprocessed, we can move on to the training of the models.

# Modelling

The models trained here are:
1. Ordinary Least Squares (OLS) Model
2. Lasso Regression Model
3. Ridge Regression Model

# Ordinary Least Squares Model

The OLS Model will be used as our baseline model.

```python
# Instantiating Linear Regression
lr = LinearRegression()
```

### Training the Model

```python
# Fitting on train dataset
lr.fit(X_train, y_train);
```

```python
# OLS Model 5-folds Cross Validation Score
lr_score = cross_val_score(lr, X_train, y_train, scoring = 'neg_mean_squared_error', cv = 5).mean()
lr_train_rmse = round(np.sqrt(-1 * lr_score), 5)
print('OLS Train RMSE is:',lr_train_rmse)
```

The scoring metric here is negative mean squared error because the scorer aims to maximize the scoring parameter. Since it is optimal to minimize RMSE, the negative equivalent would be to maximize the negative of RMSE. Details can be found in the [documentation](https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter).

```python
# Calculating R2 train score
lr_train_predict = lr.predict(X_train)
lr_train_r2 = round(metrics.r2_score(y_train, lr_train_predict), 5)
lr_train_r2
```

### Predicting against the Test Set

```python
# Predicting logged SalePrice on test data
lr_pred = lr.predict(X_test)
```

```python
# Test Score
lr_test_rmse = round(np.sqrt(metrics.mean_squared_error(lr_pred, y_test)), 5)
print('OLS Test RMSE is:',lr_test_rmse)
```

```python
# Calculating R2 test score
lr_test_r2 =round(metrics.r2_score(y_test, lr_pred), 5)
lr_test_r2
```

This is an interesting observation both the RMSE of the OLS model on both the train and test sets are very high despite the scaling of SalePrice. Given the fact that the R2 score on the test data is negative, it means that the OLS model is not able to capture information on the seen data any better than a horizontal line with the average value of all observed SalePrice (meaning of negative R2).

# Lasso Regression Model

There is an initial step of first finding the optimal alpha for the lasso regression model.

### Finding Optimal Alpha

```python
# Finding optimal alpha using LassoCV
optimal_lasso = LassoCV(cv = 5, verbose = 1)
```

```python
# Fitting on train dataset to retrieve optimal alpha
optimal_lasso.fit(X_train, y_train);
```

```python
print(optimal_lasso.alpha_)
```

### Training the Model

```python
# Instantiating Lasso regression model, using the optimal alpha
lasso = Lasso(alpha = optimal_lasso.alpha_)
```

```python
# Fitting lasso regression model with optimal alpha on train dataset
lasso.fit(X_train, y_train);
```

```python
# Lasso Model 5-folds Cross Validation Score
lasso_scores = cross_val_score(lasso, X_train, y_train, scoring = 'neg_mean_squared_error', cv = 5).mean()
lasso_train_rmse = round(np.sqrt(-1 * lasso_scores), 5)
print ("Lasso Train RMSE is:",lasso_train_rmse)
```

```python
# Calculating R2 train score
lasso_train_pred = lasso.predict(X_train)
lasso_train_r2 = round(metrics.r2_score(y_train, lasso_train_pred), 5)
lasso_train_r2
```

### Predicting against the Test Set

```python
# Predicting logged SalePrice on test data
lasso_pred = lasso.predict(X_test)
```

```python
# Test Score
lasso_test_rmse = round(np.sqrt(metrics.mean_squared_error(lasso_pred, y_test)), 5)
print('Lasso Test RMSE is:',lasso_test_rmse)
```

```python
# Calculating R2 test score
lasso_test_r2 = round(metrics.r2_score(y_test, lasso_pred), 5)
lasso_test_r2
```

```python
# Plotting test SalePrice against predicted SalePrice from Lasso regression
lasso_plot = sns.jointplot(x = y_test, y = lasso_pred)
lasso_plot.set_axis_labels('Actual Log-Transformed Sale Price', 'Predicted Log-Transformed Sale Price', fontsize = 16);

# Plotting a line of equality
lasso_plot.ax_joint.plot([10.5, 13.5], [10.5, 13.5], color = 'red', linewidth = 1, linestyle = 'dashed');
```

It can be observed that the predicted SalePrice from the Lasso model closely matches observed SalePrice. The exceptions are mainly seen from the lower log-transformed SalePrices.

# Ridge Regression

Similar to the lasso regression model, there is an initial step of finding the optimal alpha value.

### Finding Optimal Alpha

```python
# Setting range to search for optimal alpha
ridge_alphas = np.logspace(0, 5, 100)

# Finding optimal alpha using RidgeCV
optimal_ridge = RidgeCV(alphas = ridge_alphas, cv = 5)
```

```python
# Fitting on train dataset to retrieve optimal alpha
optimal_ridge.fit(X_train, y_train);
```

```python
print(optimal_ridge.alpha_)
```

### Training the Model

```python
# Instantiating Ridge regression model, using the optimal alpha
ridge = Ridge(alpha = optimal_ridge.alpha_)
```

```python
# Fitting rige regression model with optimal alpha on train dataset
ridge.fit(X_train, y_train);
```

```python
# Calculating R2 train score
ridge_train_pred = ridge.predict(X_train)
ridge_train_r2 = round(metrics.r2_score(y_train, ridge_train_pred), 5)
ridge_train_r2
```

```python
# Ridge Model 5-folds Cross Validation Score
ridge_scores = cross_val_score(ridge, X_train, y_train, scoring = 'neg_mean_squared_error', cv= 5).mean()
ridge_train_rmse = round(np.sqrt(-1 * ridge_scores), 5)
print ("Ridge Train RMSE is:",ridge_train_rmse)
```

### Predicting against the test set

```python
# Predicting log-transformed SalePrice on test data
ridge_pred = ridge.predict(X_test)
```

```python
# Calculating R2 test score
ridge_test_r2 = round(metrics.r2_score(y_test, ridge_pred), 5)
ridge_test_r2
```

```python
# Test Score
ridge_test_rmse = round(np.sqrt(metrics.mean_squared_error(ridge_pred, y_test)), 5)
print('Ridge Test RMSE is:',ridge_test_rmse)
```

```python
# Plotting test SalePrice against predicted SalePrice from Ridge regression
ridge_plot = sns.jointplot(x = y_test, y = ridge_pred)
ridge_plot.set_axis_labels('Actual Log-Transformed Sale Price', 'Predicted Log-Transformed Sale Price', fontsize = 16);

# Plotting a line of equality
ridge_plot.ax_joint.plot([10.5, 13.5], [10.5, 13.5], color = 'red', linewidth = 1, linestyle = 'dashed');
```

Similar to the observations of predicted SalePrice from the Lasso model, the predicted SalePrice from the ridge model closely follows observed SalePrice with a few exceptions in the lower SalePrice range.

# Residual Analysis

```python
# Calcuating Residuals
lasso_resid = y_test - lasso_pred
ridge_resid = y_test - ridge_pred
```

```python
# Plotting residuals of both plots
lasso_ridge_residual_plot = sns.jointplot(lasso_resid, ridge_resid)
lasso_ridge_residual_plot.set_axis_labels('Lasso Residuals', 'Ridge Residuals', fontsize=16)

# Plotting a line of equality
lasso_ridge_residual_plot.ax_joint.plot([-0.65, 0.65], [-0.65, 0.65], color = 'red', linewidth = 1, linestyle = 'dashed');
```

It is observed that residuals for both the lasso and ridge models are similar as the residuals are closely following the line of equality. Looking at the histograms on the top and right of the plot, most of the residuals are centering around the (0, 0) point.

```python
# Plotting residuals of lasso reg predictions and its residuals
lasso_residual_plot = sns.jointplot(lasso_pred, lasso_resid)
lasso_residual_plot.set_axis_labels('Predicted log-transformed Sale Price', 'Lasso Residuals', fontsize=16)

# Plotting the bands and the 0-line
lasso_residual_plot.ax_joint.plot([10.5, 13.5], [0, 0], color = 'red', linewidth = 1, linestyle = 'dashed')
lasso_residual_plot.ax_joint.plot([10.5, 13.5], [0.25, 0.25], color = 'red', linewidth = 1, linestyle = 'dashed')
lasso_residual_plot.ax_joint.plot([10.5, 13.5], [-0.25, -0.25], color = 'red', linewidth = 1, linestyle = 'dashed');
```

The residuals 'bounce' around the 0-line with no noticeable patterns, suggesting that the assumption that the relationship is linear is reasonable.

The residuals roughly form a horizontal zone around the 0-line. This suggests that the variances of the error terms are equal, albeit violated by some outliers.

There are a few residuals that stand out (as seen from those outside of the 2 outer bands. These are identified as outliers.

```python
# Plotting residuals of ridge reg predictions and its residuals
ridge_residual_plot = sns.jointplot(ridge_pred, ridge_resid)
ridge_residual_plot.set_axis_labels('Predicted log-transformed Sale Price', 'Ridge Residuals', fontsize=16)

# Plotting the bands and the 0-line
ridge_residual_plot.ax_joint.plot([10.5, 13.5], [0, 0], color = 'red', linewidth = 1, linestyle = 'dashed')
ridge_residual_plot.ax_joint.plot([10.5, 13.5], [0.25, 0.25], color = 'red', linewidth = 1, linestyle = 'dashed')
ridge_residual_plot.ax_joint.plot([10.5, 13.5], [-0.25, -0.25], color = 'red', linewidth = 1, linestyle = 'dashed');
```

The patterns and observations made in this predicted values vs residuals plot of the ridge model are similar to those of the lasso model.

To choose the best model, metrics based on the test set will be the main focus. Given the extremely large error values, OLS model will not be a candidate model.

```python
# Higher the better
print('Lasso Model test R2 is:',lasso_test_r2)
print('Ridge Model test R2 is:',ridge_test_r2)
print('\n')
# Lower the better
print('Lasso Model test RMSE is:',lasso_test_rmse)
print('Ridge Model test RMSE is:',ridge_test_rmse)
```

The lasso model has a higher test r2 score and lower rmse than the ridge model, the lasso model will be the recommended model.

# Coefficient Analysis

```python
# Create df for storing coefficients and their corresponding variable names
lasso_coefficient_df = pd.DataFrame()

# Storing all the column names of predictors
lasso_coefficient_df['Column_Name'] = X_train.columns

# Store coefficients of predictors
lasso_coefficient_df['Coefficient_Value'] = pd.Series(lasso.coef_)

# Preview variables' coefficient values
# The higher the values, the stronger effect the variables have on SalePrice
lasso_coefficient_df['Absolute_Coeff'] = lasso_coefficient_df['Coefficient_Value'].abs()
lasso_coefficient_df = lasso_coefficient_df.sort_values(by = 'Absolute_Coeff', ascending = False)
print(lasso_coefficient_df)
```

The lasso model has reduced the coefficients of many variables to 0. This means that the variables with non-zero coefficients are those that have significant influence over SalePrice. The most influential ones are those seen at the top, with the highest absolute coefficient values - Sal (Salvage Only) category of Functional (Home Functionality), Overall Quality of 2 and 9 (with negative and postive influence, respectively), Overall Condition of 3, and the list goes on.

```python
# To have a legible plot of coefficients, there is a need to filter out the weakly influential variables
lasso_coefficient_df_strong = lasso_coefficient_df.loc[lasso_coefficient_df['Absolute_Coeff'] > 0.075]
```

```python
# Plot the coefficent absolute values, in descending order
sns.set(rc = {'figure.figsize':(9,9)})
plt.barh(lasso_coefficient_df_strong["Column_Name"], lasso_coefficient_df_strong['Absolute_Coeff'].sort_values(ascending = True),
         align='center')
plt.title('Most Influential Predictors (in magnitude)');
```
